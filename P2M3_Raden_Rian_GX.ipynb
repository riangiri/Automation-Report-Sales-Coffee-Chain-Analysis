```python
# Import Libraries
from great_expectations.data_context import FileDataContext
import pandas as pd
```

```python
# buat data context
context = FileDataContext.create(project_root_dir='./')
```

```python
# Define the data source using your pandas dataframe
datasource_name = 'mycsv-cleaned'
datasource = context.sources.add_pandas(datasource_name)

# menentukan nama asset dan path untuk digunakan asset
asset_name = 'cleaned_data'
path_to_data = '/content/P2M3_Raden_Rian_data_clean.csv'  # Update the path to your CSV file
asset = datasource.add_csv_asset(asset_name, filepath_or_buffer=path_to_data)

# Build batch request
batch_request = asset.build_batch_request()
```

```python
# buat expectation suite (aturan)
# pertama di kasih nama untuk expectation
expectation_suite_name = 'expectation-clean-csv'
context.add_or_update_expectation_suite(expectation_suite_name)

# buat validator menggunakan expectation suite diatas
validator = context.get_validator(
    batch_request = batch_request,
    expectation_suite_name = expectation_suite_name
)
# Check the validator
validator.head()
```

```python
# Mencari unique value dari kolom area_code, index atau date
# Expectations 1 : to be unique
validator.expect_column_values_to_be_unique('index')
```

```python
# Membuat total_expenses tidak lebih dari 200usd
# Expectations 2 : to be between min_value and max_value
validator.expect_column_values_to_be_between(
    column='total_expenses',
    min_value=0,
    max_value=200)
```

```python
# Cek expectations type dari cofee yaitu Decaf dan Regular
# Expectations 3 : to be in set
validator.expect_column_values_to_be_in_set(
    column='type',
    value_set=['Decaf', 'Regular']
)
```

```python
# Expectations 4 : to be in type list
validator.expect_column_values_to_be_in_type_list("area_code", ["int", "int64"])
```

```python
# Expectations 5 : cek value datetime kolom date
validator.expect_column_values_to_match_strftime_format(column="date", strftime_format="%Y-%m-%d")
```

```python
# Expectations 6 : Cek jumlah data apakah diatas 1000 dengan jumlah total 1062
validator.expect_table_row_count_to_be_between(min_value=1000)
```

```python
# Expectations 7 : untuk kolom date nilainya selalu bertambah atau naik
validator.expect_column_values_to_be_increasing(column='date', parse_strings_as_datetimes=True)
```

```python
# Save into Expectation Suite
validator.save_expectation_suite(discard_failed_expectations=False)
#discard failed expectations memasukan yang berhasil saja, kalau semua pakai true
```

```python
# Buat Checkpoint
checkpoint_1 = context.add_or_update_checkpoint(
    name = 'checkpoint_1',
    validator = validator,
)
```

```python
# Run a checkpoint
checkpoint_result = checkpoint_1.run()
```
